```python
import re 

# def splitInsight(insight):
#     insight = str(insight).strip()
#     insight =re,sub(r"[--_]",'-',insgight)
#     match = re.match(r'(?:(\d+):)?(\d+):(\d+)\s*-\s(.*)',insight)
#     if match:
#         hours = int(match.group(1)) if match.group(1) else 0
#         minutes = int(match.group(2))
#         seconds = int(match.group(3))
#         description =int(match.group(4)).strip()
        
#         totalSecond = hours * 3600 + minutes + 60 + seconds
#         return pd.Series([totalSecond,description])

# return pd.Series([np.nan,none])


def splitInsight(insight):
    insight = str(insight).strip()

    # Replace double dashes/underscores with a single dash
    insight = re.sub(r"[–—_]+", '-', insight)  # fixed typo (was re,sub)

    # Match format: optional hours:minutes:seconds - description
    match = re.match(r'(?:(\d+):)?(\d+):(\d+)\s*-\s*(.*)', insight)
    
    if match:
        hours = int(match.group(1)) if match.group(1) else 0
        minutes = int(match.group(2))
        seconds = int(match.group(3))
        description = match.group(4).strip()  # was wrongly cast to int

        total_seconds = hours * 3600 + minutes * 60 + seconds
        return pd.Series([total_seconds, description])
    
    return pd.Series([np.nan, None])  # fixed typo (was `none`)


```

```python
neflix[['insight_sec', 'insight_dec']] = neflix['minute_of_life-changing_insight'].apply(splitInsight)
```

```python



def plot_boxplots(df, features):
    for feature in features:
        plt.figure(figsize=(10, 6))
        sns.boxplot(data=df, x=feature,palette=(['#d62728','#000000']))
        plt.title(f'Boxplot of {feature}', fontsize = 24)
        plt.xlabel(feature, fontsize= 16)
        plt.xticks(rotation=0)
        plt.grid()
        plt.show()

# Call the function
plot_boxplots(neflix, num_col)
```

```python
#Descriptive Statistic

#-measure of central tendency (mean,median and mode) 
#-measure of variance and standard deviation 
#-measure of frequency (frequency and range) 
#-skewness and kurtosis 
#-basoc Graphs : Bar,Line,Histogram,Pie
```

```python
numerical_col = neflix.select_dtypes(include='number')

desc_summary = pd.DataFrame({
    "mean":numerical_col.mean(),
    'median':numerical_col.median(),
    'mode':numerical_col.mode().iloc[0],
    'skewness': numerical_col.skew(),
    'kutosis': numerical_col.kurt(),
   'Range': numerical_col.max() - numerical_col.min()
    
})

desc_summary
```

```python
#Skewness is one of the basic statistic that help us to understand the data distribution and it basically give an ideal whether the data is concentrated in one side or both side or has been distributed equally in both direction


# #skweness
#  if the value = o : Not skew
# value > +0 :Right skew
# value < -o : left skew


# #Kurtosis

# -value = o : mesokurtic
# -value > 3 : leptokurtic
# -value < 3 : platykurtic
```

```python
data = neflix['insight_sec']

mean = data.mean()
std = data.std()

plt.figure(figsize=(16,4))
sns.histplot(data =data , bins=10,kde=True,color='orange', edgecolor = 'black')
plt.axvline(mean,color ='green', linestyle ='-.', linewidth = 2 ,label = 'mean')
plt.axvline(mean+std, color = 'blue', linestyle ='--',  linewidth = 2, label = '+1 std')

plt.axvline(mean - std ,color ='blue', linestyle = '--', linewidth = 2, label = '-1 std')
plt.title('insight seconds')
plt.legend()
plt.show()
            
            
            
            

```
